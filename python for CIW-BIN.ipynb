#### parsing files from CIW autoproj folder

#### keyword FROM

```python
f1 = []
new_file = []
i = 1
strip_df = pd.DataFrame(columns = ['FROM','TO'])
```

```python
from_pattern = re.compile("FROM")
```

```python
for r, d, f in os.walk(ciw_path):
    for file in f:
        #print(file)
        new_path = ciw_path + file
        with open(new_path,'r') as open_file:
            for line in open_file:
                if from_pattern.search(line):
                    #print(line)
                    from_rel = line.split("FROM", 1)
                    strip_df = strip_df.append({'FROM':from_rel[1],'TO':file}, ignore_index = True)
            print(strip_df.head(5))
        
```

```python
print(strip_df.shape)
print (type(strip_df))
print(strip_df.columns)


strip_df['FROM'] = strip_df['FROM'].replace(to_replace = "\n", value = "", regex = True)
strip_df = strip_df[strip_df['FROM'].str.strip().astype(bool)]
strip_df

strip_df = strip_df[~strip_df['FROM'].str.contains(r'[@#/,\()""*]')]
strip_df = strip_df[~strip_df['FROM'].str.contains(r"['']")]
strip_df = strip_df.replace(to_replace = ";", value = "", regex = True)
#strip_df['FROM'] = strip_df['FROM'].replace(to_replace = ".", value = "_", regex = True)
#strip_df = strip_df[strip_df[0].str.strip().astype(bool)]

strip_df.head(5)


strip_df = strip_df.apply(lambda x:x.astype(str).str.upper())

strip_df.reset_index(drop=True, inplace=True)
#strip_df = strip_df.set_index('index')

print(strip_df.shape)
strip_df = strip_df.drop_duplicates()
print(strip_df.shape)

strip_df['FROM'] = strip_df['FROM'].str.strip()

strip_df['FROM'] = strip_df['FROM'].str.split(" ").str[0]

strip_df.reset_index(drop=True, inplace=True)
#strip_df = strip_df.set_index('index')
print(strip_df.shape)
strip_df = strip_df.drop_duplicates()
print(strip_df.shape)

strip_df = strip_df.replace(to_replace = "_", value = ".", regex = True)

strip_df
```

```python
strip_df.to_csv('FROM RELATIONSHIP.csv')
```
